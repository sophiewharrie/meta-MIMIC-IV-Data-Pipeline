Use frequency-based filtering, where the least frequent codes are excluded. Using arbitrary thresholds for now, adjust later with inupt from expert domain knowledge

```python
# diagosis features
feat = 'diag'
featsum_df = pd.read_csv(f'data/summary/{feat}_summary.csv')
featsum_df = featsum_df.sort_values(by='total_count', ascending=False)
threshold_count = featsum_df['total_count'].quantile(0.7)
threshold_freq = featsum_df['mean_frequency'].quantile(0.7)
featsum_df = featsum_df[(featsum_df['total_count']>=threshold_count)&(featsum_df['mean_frequency']>=threshold_freq)]
# save the new feature file (first column of the reduced summary)
featsum_df.iloc[:,:1].to_csv(f'data/summary/{feat}_features.csv', index=None)
len(featsum_df)
```

```python
# medication features
feat = 'med'
featsum_df = pd.read_csv(f'data/summary/{feat}_summary.csv')
featsum_df = featsum_df.sort_values(by='total_count', ascending=False) 
threshold_count = featsum_df['total_count'].quantile(0.5)
threshold_freq = featsum_df['mean_frequency'].quantile(0.5)
featsum_df = featsum_df[(featsum_df['total_count']>=threshold_count)&(featsum_df['mean_frequency']>=threshold_freq)]
# save the new feature file (first column of the reduced summary)
featsum_df.iloc[:,:1].to_csv(f'data/summary/{feat}_features.csv', index=None)
len(featsum_df)
```

```python
# chart features
feat = 'chart'
featsum_df = pd.read_csv(f'data/summary/{feat}_summary.csv')
featsum_df = featsum_df.sort_values(by='total_count', ascending=False)
threshold_count = featsum_df['total_count'].quantile(0.7)
threshold_freq = featsum_df['mean_frequency'].quantile(0.7)
featsum_df['missing_to_total_count'] = featsum_df['missing_count'] / featsum_df['total_count']
threshold_missing = featsum_df['missing_to_total_count'].quantile(0.8)
featsum_df = featsum_df[(featsum_df['total_count']>=threshold_count)&(featsum_df['mean_frequency']>=threshold_freq)&(featsum_df['missing_to_total_count']<=threshold_missing)]
# save the new feature file (first column of the reduced summary)
featsum_df.iloc[:,:1].to_csv(f'data/summary/{feat}_features.csv', index=None)
len(featsum_df)
```

## 7. Time-Series Representation
In this section, please choose how you want to process and represent time-series data.

- First option is to select the length of time-series data you want to include for this study. (Default is 72 hours)

- Second option is to select bucket size which tells in what size time windows you want to divide your time-series.<br>
For example, if you select **2** bucket size, it wil aggregate data for every 2 hours and <br>a time-series of length 24 hours will be represented as time-series with 12 time-windows <br>where data for every 2 hours is agggregated from original raw time-series.

During this step, we will also save the time-series data in data dictionaries in the format that can be directly used for following deep learning analysis.

### Imputation
You can also choose if you want to impute lab/chart values. The imputation will be done by froward fill and mean or median imputation.<br>
Values will be forward fill first and if no value exists for that admission we will use mean or median value for the patient.

The data dictionaries will be saved in **./data/dict/**

## 8. Meta-learning dataset

For a meta-learning dataset, we need to create
- Combined dataset of all tasks, containing the following columns:
    - Feature columns with prefix **X**
    - Label column named **Y**
    - **task** index column
    - **task_train** column for support/query split (earlier/later samples)
- Fix the number of samples per task
- Intervention mask, with the same column names as the main dataset, but values are binary and intervention values are marked with 1

```python
import pandas as pd
from tqdm import tqdm
import glob
```

```python
num_samples = []

all_data = pd.DataFrame()
all_patients = [int(s.split('/')[2]) for s in glob.glob('data/csv/*/')]
patient_task_map = dict(zip(all_patients, range(len(all_patients))))

for patient in tqdm(all_patients):
    dynamic_data_path = 'data/csv/{}/dynamic.csv'
    static_data_path = 'data/csv/{}/static.csv'
    demo_data_path = 'data/csv/{}/demo.csv'

    # get patient data
    dynamic_df = pd.read_csv(dynamic_data_path.format(patient))
    static_df = pd.read_csv(static_data_path.format(patient))
    demo_df = pd.read_csv(demo_data_path.format(patient))[['Age','gender']]

    # combine header and 1st row into a single header
    static_df.columns = 'X_' + static_df.columns + '_' + static_df.iloc[0].astype(str)
    static_df = static_df.drop(0).reset_index().drop(columns='index')
    dynamic_df.columns = 'X_' + dynamic_df.columns + '_' + dynamic_df.iloc[0].astype(str)
    dynamic_df = dynamic_df.drop(0).reset_index().drop(columns='index')

    # convert gender to integer
    demo_df['gender'] = demo_df['gender'].map({'F':1, 'M':0})
    demo_df = demo_df.rename(columns={'Age':'X_Age','gender':'X_Gender'}) 

    # combine static and demo data with dynamic data
    patient_df = pd.concat([dynamic_df, static_df, demo_df], axis=1).fillna(method='ffill')
    patient_df = patient_df.assign(task=patient_task_map[patient])

    num_samples.append({'id':patient, 'num_samples':len(patient_df)})
    # only consider first 30 hours of data, first 12 hours are support data
    patient_df = patient_df[0:30]
    patient_df['task_train'] = [1 if i < 12 else 0 for i in range(30)]
    all_data = pd.concat([all_data, patient_df])
```

```python
all_data.to_csv('data/meta/raw_data.csv', index=None) # save checkpoint here, in case want to change Y later
```

```python
# save the patient task mapping
pd.DataFrame({'task':patient_task_map.values(), 'id':patient_task_map.keys()}).to_csv('data/meta/task_ids.csv', index=None)
```

```python
# pick a label for prediction - select a chart event and predict x-hours ahead
# (can't use default task because it's predicting 30-day outcomes (fixed for all timestamps), so is not suitable for the meta-learning setup)

variable_to_predict = 'X_CHART.3_220181.0' # Non Invasive Blood Pressure mean
x_hours = 6

all_data['Y'] = all_data.groupby('task')[variable_to_predict].shift(-x_hours)
all_data = all_data.drop(columns=variable_to_predict)
all_data = all_data.dropna()
```

```python
# remove variables closely correlated with Y
from sklearn.feature_selection import r_regression

threshold = 0.5 # anything with pearson correlation higher will be removed
correlations = r_regression(all_data.drop(columns=['task','task_train','Y']), all_data['Y'])

tmp = pd.DataFrame({'key':all_data.drop(columns=['task','task_train','Y']).keys(),'corr':np.abs(correlations)}).sort_values(by='corr', ascending=False)
to_remove = tmp[tmp['corr']>=threshold]['key'].tolist()
print('Removing...')
for k in to_remove: print('{}'.format(k))
all_data = all_data.drop(columns=to_remove)
```

```python
# remove patients with Y == 0 (indicates missing values) COMMENT OUT IF NOT NECESSARY
to_remove = all_data[all_data['Y']==0]['task'].unique()
all_data = all_data[~all_data['task'].isin(to_remove)]

# re-index the task numbers
patient_map = pd.DataFrame({'old_task':patient_task_map.values(),'id':patient_task_map.keys()})
patient_map = patient_map[~patient_map['old_task'].isin(to_remove)]
patient_map['task'] = range(len(patient_map))
updated_map = dict(zip(patient_map['old_task'], patient_map['task']))
all_data['task'] = all_data['task'].map(updated_map)
```

```python
# update the saved patient task mapping
patient_map = patient_map.drop(columns='old_task')
patient_map.to_csv('data/meta/task_ids.csv', index=None)
```

```python
# save final dataset for meta-learning
all_data.to_csv('data/meta/final_data.csv', index=None)
```

```python
# get detailed descriptions of all columns in the dataset

key_data = []
for k in all_data.keys():
    tmp = {'column_name':k, 'itemid':k.split('_')[-1]}
    tmp['type'] = k.split('_')[1].split('.')[0] if len(k.split('_'))>1 else k
    key_data.append(tmp)
    
key_data = pd.DataFrame(key_data)

# map the feature names
chart_map_df = pd.read_csv('../mimiciv/2.0/icu/d_items.csv.gz', compression='gzip')
chart_map_df = chart_map_df[chart_map_df['linksto']=='chartevents'].drop(columns=['abbreviation','linksto'])
chart_map_df['type'] = 'CHART'
chart_map_df['itemid'] = chart_map_df['itemid'].astype(float).astype(str)
icd_map_df = pd.read_csv('../mimiciv/2.0/hosp/d_icd_diagnoses.csv.gz', compression='gzip').rename(columns={'icd_code':'itemid','long_title':'label'})
icd_map_df = icd_map_df[icd_map_df['icd_version']==10]
icd_map_df['type'] = 'COND'
med_map_df = pd.read_csv('../mimiciv/2.0/icu/d_items.csv.gz', compression='gzip')
med_map_df = med_map_df[med_map_df['linksto']=='inputevents'].drop(columns=['abbreviation','linksto'])
med_map_df['type'] = 'MEDS'
med_map_df['itemid'] = med_map_df['itemid'].astype(float).astype(str)

map_df = pd.concat([chart_map_df, icd_map_df, med_map_df])

key_data = pd.merge(key_data, map_df, on=['itemid','type'], how='left')
key_data.to_csv('data/meta/column_descriptions.csv', index=None)
```

```python
# make intervention mask
interv_mask = all_data.copy()

for k in tqdm(interv_mask.keys()):
    if k.startswith('X_MEDS'):
        interv_mask[k] = (interv_mask[k]!=0).astype(int)
    else:
        interv_mask[k] = 0

interv_mask.to_csv('data/meta/interv_mask.csv', index=None)
```
